## Importing necessary libraries

```python
import os
import sys
import warnings

import numpy as np
import pandas as pd
import matplotlib.font_manager
from mizani.transforms import log_trans
from mizani.formatters import percent_format
from mizani.formatters import log_format
from plotnine import *
from datetime import datetime
from scipy.stats import norm
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer

warnings.filterwarnings("ignore")
```

## Dataset cleaning and creation of filtered table

```python
# Import and clean dataset
df = pd.read_csv('morg-2014-emp.csv')

# Flag with '1' where the occupation code if for people in Education Training and Library
df['occ'] = np.where(df['occ2012'].between(2200, 2550), 1, 0)

# Take 'grade92' below 43, this will be further elaborated on in the next section
df = df[df['grade92'] >= 43]

# Occupation applied filter named as 'filtered_data'
filtered_data = df[df['occ'].isin([1])]
```

### <span style="color:red">Please note 'edu_trai_lib' denotes Education, Training and Library.</span>

## Creation of additional variables

```python
# Generate variables for 'df'
filtered_data['female'] = (filtered_data['sex'] == 2).astype(int)

# Hourly wages calculated and stored
filtered_data['hourly_wage'] = filtered_data['earnwke'] / filtered_data['uhours']

# Drop extreme values that are below 1 (hourly wages less than 1 USD per hour may be an error)
filtered_data = filtered_data[filtered_data['hourly_wage'] >= 1]

# Log of hourly wages
filtered_data['ln_hourly_wage'] = np.log(filtered_data['hourly_wage'])

# Different levels of education stored based on their codes

filtered_data['BA_degree'] = (filtered_data['grade92'] == 43).astype(int)
filtered_data['MA_degree'] = (filtered_data['grade92'] == 44).astype(int)
filtered_data['Prof_degree'] = (filtered_data['grade92'] == 45).astype(int)
filtered_data['PhD_degree'] = (filtered_data['grade92'] == 46).astype(int)
```

## Descriptive statistics for 'filtered_data'

```python
# Descriptive statistics for 'filtered_data'
desc_stats_filtered = filtered_data[['hourly_wage', 'age']].describe(percentiles=[.05, .95])

# Print Descriptive Statistics
desc_stats_filtered
```

## Creating 3 models using OLS

### Model 1: Hourly wage gap for women in edu_trai_lib

```python
# Model 1: hourly wage gap for women in edu_trai_lib
model1 = smf.ols(formula="ln_hourly_wage ~ female", data=filtered_data).fit()
model1.summary()
```

### Model 2: Hourly wage gap for women in edu_trai_lib who are union members

```python
# Model 2: Hourly wage gap for women in edu_trai_lib who are union members
model2 = smf.ols(formula="ln_hourly_wage ~ female + unionmme", data=filtered_data).fit()
model2.summary()
```

### Model 3: Hourly wage gap for edu_trai_lib subset including unionship and education levels

```python
# Model 3: Hourly wage gap for edu_trai_lib subset including unionship and education levels
model3 = smf.ols(formula="ln_hourly_wage ~ female + unionmme + BA_degree + MA_degree + Prof_degree + PhD_degree", data=filtered_data).fit()
model3.summary()
```

```python
stargazer = Stargazer([model1, model2, model3])
stargazer.covariate_order(["female", "unionmme[T.Yes]", "BA_degree", "MA_degree", "PhD_degree", "Prof_degree", "Intercept"])
stargazer
```

## Creating graphs using ggplot

### Graph 1: Wage distribution for 'all_data'

```python
# Boxplot of hourly wages by gender for Education Related Jobs
boxplot_data = (
    ggplot(data=filtered_data, mapping=aes(x='factor(female)', y='hourly_wage')) +
    geom_boxplot(fill=['pink', 'lightblue']) +
    scale_y_log10() +
    labs(title='Hourly Wage Distribution by Gender in edu_trai_lib', x='Gender (0: Male, 1: Female)', 
         y='Hourly Wage (Log Scale)')
)

boxplot_data
```

### Graph 2: Wages by Unionship for 'filtered_data'

```python
# Boxplot of hourly wages by Unionship for Education Related Jobs

filtered_data_g2 = filtered_data[filtered_data['sex'].isin([1])]

boxplot_education_data = (
    ggplot(data=filtered_data_g2, mapping=aes(x='unionmme', y='hourly_wage')) +
    geom_boxplot(fill=['pink', 'lightblue']) +
    scale_y_log10() +
    labs(title='Hourly Wage Distribution by Unionship For edu_trai_lib Jobs', x='Unionship', 
         y='Hourly Wage (Log Scale)')
)

boxplot_education_data
```
